```python
# Optimized SisFall pre-fall detection pipeline (copy-paste to Kaggle/Colab)
# - Dynamic pre-fall labeling based on signal energy
# - Butterworth cutoff raised to 30 Hz
# - Attention-augmented BiLSTM
# - Per-class focal loss + class_weight balancing
# - Consecutive-window alerting for simulation
# - Save model and scaler at end
# NOTE: Adjust DATA_ROOT if needed.

import os, re, pickle, math
import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.utils.class_weight import compute_class_weight

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Bidirectional, LSTM, Dense, Dropout, BatchNormalization, Concatenate, Attention
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
import tensorflow.keras.backend as K

from scipy.signal import butter, filtfilt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,5)

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# -------------------------
# STEP 0: Configure dataset path
# -------------------------
DATA_ROOT = "/kaggle/input/sisfall/SisFall_dataset"  # adjust if needed
assert os.path.exists(DATA_ROOT), f"Dataset path not found: {DATA_ROOT}"
print("Dataset path OK:", DATA_ROOT)

# -------------------------
# STEP 1: File discovery & parser
# -------------------------
pattern = re.compile(r"(?P<code>[FD]\d{2})_(?P<subject>S[AE]\d{2})_R(?P<trial>\d+)\.txt", re.I)

def parse_txt_file(path):
    """Read a SisFall text file into numpy array (n_samples, 9)."""
    rows = []
    with open(path, "r", errors="ignore") as f:
        for line in f:
            line = line.strip().rstrip(";")
            if not line:
                continue
            toks = [t.strip() for t in line.split(",") if t.strip() != ""]
            if len(toks) != 9:
                continue
            try:
                vals = [float(x) for x in toks]
            except:
                continue
            rows.append(vals)
    if len(rows) == 0:
        return np.empty((0,0), dtype=np.float32)
    return np.array(rows, dtype=np.float32)

records = []
for root, dirs, files in os.walk(DATA_ROOT):
    for fname in files:
        if not fname.lower().endswith(".txt"):
            continue
        m = pattern.search(fname)
        if not m:
            continue
        g = m.groupdict()
        code = g["code"].upper()
        subject = g["subject"].upper()
        label = "fall" if code.startswith("F") else "normal"
        records.append({
            "path": os.path.join(root, fname),
            "fname": fname,
            "code": code,
            "subject": subject,
            "label": label
        })

df_files = pd.DataFrame(records)
print("Files found:", len(df_files))
print(df_files.label.value_counts())

# -------------------------
# STEP 2: Load trials into memory
# -------------------------
all_trials = []
for _, row in tqdm(df_files.iterrows(), total=len(df_files), desc="Loading files"):
    arr = parse_txt_file(row["path"])
    if arr.ndim != 2 or arr.shape[1] != 9:
        continue
    all_trials.append({
        "data": arr,
        "label": row["label"],
        "subject": row["subject"],
        "fname": row["fname"]
    })
print("Trials loaded:", len(all_trials))

# -------------------------
# STEP 3: Filtering & windowing helpers
# -------------------------
def butter_lowpass_filter(data, fs=200, cutoff=30, order=4):
    """Zero-phase Butterworth lowpass filter along time axis (axis=0)."""
    if data.shape[0] < (order * 3):
        return data
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    try:
        return filtfilt(b, a, data, axis=0)
    except Exception:
        return data

def create_windows(trial_array, window_size=200, step_size=100, fs=200, apply_filter=True):
    """Return windows of shape (n_windows, window_size, n_channels)."""
    if apply_filter:
        trial_array = butter_lowpass_filter(trial_array, fs=fs, cutoff=30)
    n = trial_array.shape[0]
    windows = []
    for start in range(0, n - window_size + 1, step_size):
        windows.append(trial_array[start:start + window_size])
    if len(windows) == 0:
        return np.empty((0, window_size, trial_array.shape[1]))
    return np.stack(windows, axis=0)

# -------------------------
# STEP 4: Dynamic labeling helper (data-driven pre-fall window selection)
# -------------------------
def dynamic_label_windows(trial_array, window_size=200, step_size=100, fs=200, prefall_seconds=2.0):
    """
    For fall trials: detect impact via smoothed acceleration magnitude peak and
    label windows:
      2 = fall (impact windows near peak)
      1 = pre-fall (approx prefall_seconds before impact)
      0 = normal
    For non-fall trials: all windows labeled 0.
    Returns: wins (n_windows, window_size, nchan), labels (n_windows,)
    """
    wins = create_windows(trial_array, window_size=window_size, step_size=step_size, fs=fs, apply_filter=True)
    n_windows = wins.shape[0]
    if n_windows == 0:
        return wins, np.empty((0,), dtype=int)

    # use acceleration channels for energy: assume channels 0-2 are accelerometer X,Y,Z
    accel = trial_array[:, :3]
    accel_mag = np.linalg.norm(accel, axis=1)  # magnitude
    # Smooth magnitude with 0.2-1.0s window (use 0.5s here)
    smooth_win = int(0.5 * fs)
    if smooth_win < 1:
        smooth_win = 1
    kernel = np.ones(smooth_win) / smooth_win
    energy = np.convolve(accel_mag, kernel, mode='same')

    # Peak index of energy
    impact_idx = int(np.argmax(energy))
    # Convert prefall seconds to index
    prefall_idx = max(impact_idx - int(prefall_seconds * fs), 0)

    labels = np.zeros(n_windows, dtype=int)
    for i in range(n_windows):
        window_start_idx = i * step_size
        window_end_idx = window_start_idx + window_size - 1
        # If window overlaps the impact region (impact_idx Â± small margin), mark as fall impact
        if window_start_idx <= impact_idx <= window_end_idx:
            labels[i] = 2
        # if window is within prefall region (start >= prefall_idx and ends before impact)
        elif window_start_idx >= prefall_idx and window_end_idx < impact_idx:
            labels[i] = 1
        else:
            labels[i] = 0
    return wins, labels

# -------------------------
# STEP 5: Build windows & labels for entire dataset
# -------------------------
WINDOW = 200   # 1 second windows at 200 Hz
STEP = 100     # 0.5 second hop
PREFALL_SECONDS = 2.0

X_list, y_list, subj_list, meta_list = [], [], [], []

for t in tqdm(all_trials, desc="Windowing & dynamic labelling"):
    arr = t["data"]
    if t["label"] == "fall":
        wins, labels = dynamic_label_windows(arr, window_size=WINDOW, step_size=STEP, fs=200, prefall_seconds=PREFALL_SECONDS)
    else:
        wins = create_windows(arr, window_size=WINDOW, step_size=STEP, apply_filter=True)
        labels = np.zeros(wins.shape[0], dtype=int) if wins.shape[0] > 0 else np.empty((0,), dtype=int)

    if wins.shape[0] == 0:
        continue

    for i in range(wins.shape[0]):
        X_list.append(wins[i])
        y_list.append(int(labels[i]))
        subj_list.append(t["subject"])
        meta_list.append(t["fname"])

X = np.array(X_list)             # shape (N_windows, WINDOW, n_channels)
y = np.array(y_list, dtype=int)
subjects = np.array(subj_list)
meta = np.array(meta_list)

print("Windows created:", X.shape)
print("Label counts (0=normal,1=pre-fall,2=fall):", np.bincount(y))

# -------------------------
# STEP 6: Grouped train/test split
# -------------------------
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=SEED)
train_idx, test_idx = next(gss.split(X, y, groups=subjects))
X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]
subjects_train = subjects[train_idx]
subjects_test = subjects[test_idx]
meta_test = meta[test_idx]

print("Train windows:", X_train.shape, "Test windows:", X_test.shape)
print("Train label counts:", np.bincount(y_train), "Test label counts:", np.bincount(y_test))

# -------------------------
# STEP 7: Standardize channels (fit on training set)
# -------------------------
n_train = X_train.shape[0]
_, wsize, nch = X_train.shape

scaler = StandardScaler()
X_train_flat = X_train.reshape(n_train, -1)
X_train_scaled = scaler.fit_transform(X_train_flat).reshape(n_train, wsize, nch)

n_test = X_test.shape[0]
X_test_flat = X_test.reshape(n_test, -1)
X_test_scaled = scaler.transform(X_test_flat).reshape(n_test, wsize, nch)

X_train = X_train_scaled
X_test = X_test_scaled

print("Standardization done. Example train shape:", X_train.shape)

# -------------------------
# STEP 8: Class weights (avoid destructive oversampling)
# -------------------------
classes = np.unique(y_train)
class_weights_list = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = {int(c): float(w) for c, w in zip(classes, class_weights_list)}
print("Class weights:", class_weights)

# -------------------------
# STEP 9: Focal loss (per-class alpha)
# -------------------------
def focal_loss(gamma=2.0, alpha=[0.5, 1.0, 1.5]):
    """
    y_true: integer labels (batch,)
    y_pred: probabilities (batch, n_classes)
    alpha: list or array of length n_classes (higher for rarer classes)
    """
    alpha = tf.constant(alpha, dtype=tf.float32)
    def loss_fn(y_true, y_pred):
        # y_true may be shape (batch,) or (batch,1)
        y_true_flat = tf.cast(tf.reshape(y_true, (-1,)), tf.int32)
        y_true_o = tf.one_hot(y_true_flat, depth=tf.shape(y_pred)[-1])
        y_pred = tf.clip_by_value(y_pred, 1e-9, 1.0)
        p_t = tf.reduce_sum(y_true_o * y_pred, axis=-1)
        alpha_factor = tf.reduce_sum(y_true_o * alpha, axis=-1)
        modulating_factor = tf.pow(1.0 - p_t, gamma)
        loss = - alpha_factor * modulating_factor * tf.math.log(p_t)
        return tf.reduce_mean(loss)
    return loss_fn

# -------------------------
# STEP 10: Model (Attention + BiLSTM)
# -------------------------
def build_model(input_shape):
    inputs = Input(shape=input_shape)
    x = Bidirectional(LSTM(128, return_sequences=True))(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    # Self-attention: query=key=value -> temporal attention
    attn = Attention()([x, x])
    x = Concatenate()([x, attn])

    x = Bidirectional(LSTM(64, return_sequences=False))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    x = Dense(128, activation="relu")(x)
    x = Dropout(0.4)(x)

    outputs = Dense(3, activation="softmax")(x)
    model = Model(inputs, outputs)
    model.compile(
        optimizer=Adam(learning_rate=3e-4),
        loss=focal_loss(gamma=2.0, alpha=[0.5, 1.0, 1.5]),
        metrics=["accuracy"]
    )
    return model

model = build_model((WINDOW, nch))
model.summary()

# -------------------------
# STEP 11: Training
# -------------------------
es = EarlyStopping(monitor="val_loss", patience=12, restore_best_weights=True, verbose=1)
rlr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, verbose=1, min_lr=1e-6)

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=80,
    batch_size=64,
    class_weight=class_weights,
    callbacks=[es, rlr],
    verbose=1
)

# -------------------------
# STEP 12: Evaluation
# -------------------------
y_pred_prob = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_prob, axis=1)

test_acc = accuracy_score(y_test, y_pred)
prec_prefall = precision_score(y_test, y_pred, labels=[1], average='macro', zero_division=0)
rec_prefall = recall_score(y_test, y_pred, labels=[1], average='macro', zero_division=0)
print("Test Accuracy:", test_acc)
print(f"Pre-fall precision: {prec_prefall:.4f}, recall: {rec_prefall:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=["normal", "pre-fall", "fall"], zero_division=0))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='magma',
            xticklabels=["normal","pre-fall","fall"],
            yticklabels=["normal","pre-fall","fall"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()

# Normalized confusion
cm_norm = cm.astype('float') / (cm.sum(axis=1, keepdims=True) + 1e-9)
plt.figure(figsize=(6,5))
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='crest',
            xticklabels=["normal","pre-fall","fall"], yticklabels=["normal","pre-fall","fall"])
plt.title("Confusion Matrix (normalized)")
plt.show()

# Plot training curves
def plot_history(history):
    fig, ax = plt.subplots(1,2, figsize=(14,4))
    ax[0].plot(history.history['loss'], label='train loss')
    ax[0].plot(history.history['val_loss'], label='val loss')
    ax[0].set_title('Loss'); ax[0].legend(); ax[0].grid(True)

    ax[1].plot(history.history['accuracy'], label='train acc')
    ax[1].plot(history.history['val_accuracy'], label='val acc')
    ax[1].set_title('Accuracy'); ax[1].legend(); ax[1].grid(True)
    plt.show()

plot_history(history)

# -------------------------
# STEP 13: Simulate streaming on a fall trial (consecutive-window alerting)
# -------------------------
fall_trials = [t for t in all_trials if t["label"] == "fall"]

def simulate_stream_from_trial_array(trial_array, trial_name="trial", fs=200, pref_thresh=0.45, fall_thresh=0.65, consecutive=3, visualize=True):
    wins = create_windows(trial_array, window_size=WINDOW, step_size=STEP, apply_filter=True)
    if wins.shape[0] == 0:
        print("No windows to simulate.")
        return
    times = np.arange(wins.shape[0]) * (STEP / fs)
    probs = []
    pre_count = 0
    fall_count = 0
    alerts = []
    for i, w in enumerate(wins):
        w_scaled = scaler.transform(w.reshape(1, -1)).reshape(1, WINDOW, nch)
        p = model.predict(w_scaled, verbose=0)[0]
        probs.append(p)
        if p[2] >= fall_thresh:
            fall_count += 1
            pre_count = 0
            if fall_count >= consecutive:
                alerts.append(("FALL", times[i], float(p[2])))
                print(f"[ALERT] FALL at ~{times[i]:.2f}s  p_fall={p[2]:.3f}")
        elif p[1] >= pref_thresh:
            pre_count += 1
            fall_count = 0
            if pre_count >= consecutive:
                alerts.append(("PRE-FALL", times[i], float(p[1])))
                print(f"[WARNING] PRE-FALL at ~{times[i]:.2f}s  p_prefall={p[1]:.3f}")
        else:
            pre_count = 0
            fall_count = 0

    probs = np.array(probs)
    if visualize:
        plt.figure(figsize=(12,4))
        plt.plot(times, probs[:,0], label="Normal", lw=2)
        plt.plot(times, probs[:,1], label="Pre-Fall", lw=2)
        plt.plot(times, probs[:,2], label="Fall", lw=2)
        plt.xlabel("Time (s)"); plt.ylabel("Probability")
        plt.title(f"Predicted probabilities â {trial_name}")
        plt.legend(); plt.grid(True); plt.show()
    return alerts, probs, times

if len(fall_trials) > 0:
    print("Simulating and visualizing for:", fall_trials[0]["fname"])
    simulate_stream_from_trial_array(fall_trials[0]["data"], trial_name=fall_trials[0]["fname"])
else:
    print("No fall trials available to visualize.")

# -------------------------
# STEP 14: Compute lead time statistics for detected pre-fall on test set
# -------------------------
def compute_lead_times_on_test(X_test, y_test, meta_test, fs=200, pref_thresh=0.45, consecutive=3):
    """
    For each true fall trial window in test set, try to find earliest pre-fall detection
    and compute time before impact (lead time). We need mapping windows -> trial timeline,
    but we only have windows in test set (some trials split across train/test by subject grouping).
    This function is an approximate per-window lead-time estimator.
    """
    # Group windows by meta (filename)
    df = pd.DataFrame({"meta": meta_test, "y_true": y_test})
    unique_meta = df['meta'].unique()
    lead_times = []
    for m in unique_meta:
        mask = (meta_test == m)
        if not np.any(mask):
            continue
        windows = X_test[mask]
        truths = y_test[mask]
        # skip if trial not a fall trial (no ground-truth fall windows)
        if not np.any(truths == 2):
            continue
        # simulate stream until first PRE-FALL or FALL detection
        pre_count = fall_count = 0
        detected_time = None
        for i, w in enumerate(windows):
            w_scaled = scaler.transform(w.reshape(1, -1)).reshape(1, WINDOW, nch)
            p = model.predict(w_scaled, verbose=0)[0]
            t = i * (STEP / fs)
            if p[2] >= 0.65:
                fall_count += 1
                pre_count = 0
                if fall_count >= consecutive:
                    detected_time = t
                    break
            elif p[1] >= pref_thresh:
                pre_count += 1
                fall_count = 0
                if pre_count >= consecutive:
                    detected_time = t
                    break
            else:
                pre_count = fall_count = 0
        # Find true impact time from truths: first window labeled 2
        impact_idx_local = np.where(truths == 2)[0]
        if impact_idx_local.size == 0:
            continue
        impact_time = impact_idx_local[0] * (STEP / fs)
        if detected_time is not None:
            lead = impact_time - detected_time
            lead_times.append(lead)
    if len(lead_times) == 0:
        print("No lead times computed (no detections).")
        return None
    lead_times = np.array(lead_times)
    print(f"Lead time stats (s): mean={lead_times.mean():.3f}, median={np.median(lead_times):.3f}, std={lead_times.std():.3f}, n={len(lead_times)}")
    return lead_times

compute_lead_times_on_test(X_test, y_test, meta_test)

# -------------------------
# STEP 15: Save model & scaler
# -------------------------
MODEL_NAME = "bilstm_attention_prefall.h5"
SCALER_NAME = "scaler_prefall.pkl"
model.save(MODEL_NAME)
with open(SCALER_NAME, "wb") as f:
    pickle.dump(scaler, f)
print("Model saved to", MODEL_NAME)
print("Scaler saved to", SCALER_NAME)
```

```python
# ====== Visualization & SHAP cell (paste below your training cell) ======
import os, random, pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score
from tensorflow.keras.models import load_model
import tensorflow as tf

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,5)

# Try to reuse variables from previous cell; if missing, attempt to load saved artifacts
MODEL_NAME = "bilstm_attention_prefall.h5"
SCALER_NAME = "scaler_prefall.pkl"

# helper - check existence
try:
    _ = model  # if this fails, model not in namespace
    print("Using model from environment.")
except Exception:
    if os.path.exists(MODEL_NAME):
        print("Loading model from file:", MODEL_NAME)
        # define focal_loss same as training to load model
        def focal_loss(gamma=2.0, alpha=[0.5, 1.0, 1.5]):
            alpha = tf.constant(alpha, dtype=tf.float32)
            def loss_fn(y_true, y_pred):
                y_true_flat = tf.cast(tf.reshape(y_true, (-1,)), tf.int32)
                y_true_o = tf.one_hot(y_true_flat, depth=tf.shape(y_pred)[-1])
                y_pred = tf.clip_by_value(y_pred, 1e-9, 1.0)
                p_t = tf.reduce_sum(y_true_o * y_pred, axis=-1)
                alpha_factor = tf.reduce_sum(y_true_o * alpha, axis=-1)
                modulating_factor = tf.pow(1.0 - p_t, gamma)
                loss = - alpha_factor * modulating_factor * tf.math.log(p_t)
                return tf.reduce_mean(loss)
            return loss_fn
        try:
            model = load_model(MODEL_NAME, custom_objects={"focal_loss": focal_loss})
        except Exception:
            model = load_model(MODEL_NAME, compile=False)
            model.compile(optimizer=tf.keras.optimizers.Adam(3e-4),
                          loss=focal_loss(gamma=2.0, alpha=[0.5,1.0,1.5]),
                          metrics=["accuracy"])
        print("Model loaded.")
    else:
        raise FileNotFoundError(f"Model variable not found and model file {MODEL_NAME} does not exist.")

try:
    _ = scaler
    print("Using scaler from environment.")
except Exception:
    if os.path.exists(SCALER_NAME):
        with open(SCALER_NAME, "rb") as f:
            scaler = pickle.load(f)
        print("Scaler loaded from", SCALER_NAME)
    elif os.path.exists("scaler.pkl"):
        with open("scaler.pkl", "rb") as f:
            scaler = pickle.load(f)
        print("Loaded scaler.pkl")
    else:
        raise FileNotFoundError("Scaler not found in environment and no scaler file found.")

# ensure X, y, X_test, y_test exist - otherwise try to rebuild quickly (requires create_windows & all_trials)
if 'X' not in globals() or 'y' not in globals():
    raise RuntimeError("X and y windows not found in current environment. Please run the windowing cell first (the training cell).")

# flatten helper shapes
N_CH = nch if 'nch' in globals() else X.shape[2]
print("Window shape:", X.shape, "Channels:", N_CH)

# -----------------------
# 1) Dataset label distribution
# -----------------------
plt.figure(figsize=(6,4))
sns.countplot(x=y)
plt.xticks([0,1,2], ["normal","pre-fall","fall"])
plt.title("Label distribution (all windows)")
plt.show()

# -----------------------
# 2) Raw trial time-series plots (accel + gyro channels)
# -----------------------
def plot_raw_trial(trial_array, title="trial", channels_to_plot=None, fs=200, figsize=(14,6)):
    if channels_to_plot is None:
        channels_to_plot = [0,1,2,3,4,5]  # show accel & gyro if present
    t = np.arange(trial_array.shape[0]) / fs
    n = len(channels_to_plot)
    fig, axs = plt.subplots(n, 1, figsize=(figsize[0], 2.0*n), sharex=True)
    if n == 1:
        axs = [axs]
    for i, ch in enumerate(channels_to_plot):
        axs[i].plot(t, trial_array[:,ch])
        axs[i].set_ylabel(f"ch{ch}")
        axs[i].grid(True)
    axs[-1].set_xlabel("Time (s)")
    fig.suptitle(title)
    plt.show()

# pick one fall and one normal trial if available
fall_trials = [t for t in all_trials if t["label"] == "fall"]
normal_trials = [t for t in all_trials if t["label"] == "normal"]

if len(fall_trials) > 0:
    print("Raw signals for fall trial:", fall_trials[0]["fname"])
    plot_raw_trial(fall_trials[0]["data"], title=fall_trials[0]["fname"])
if len(normal_trials) > 0:
    print("Raw signals for normal trial:", normal_trials[0]["fname"])
    plot_raw_trial(normal_trials[0]["data"], title=normal_trials[0]["fname"])

# -----------------------
# 3) Example windows for each class (normal / pre-fall / fall)
# -----------------------
def plot_window_example(window_arr, class_name="class", fs=200):
    t = np.linspace(0, window_arr.shape[0]/fs, window_arr.shape[0])
    plt.figure(figsize=(12,4))
    for ch in range(min(6, window_arr.shape[1])):  # plot first 6 channels
        plt.plot(t, window_arr[:,ch], label=f"ch{ch}")
    plt.xlabel("Time (s)"); plt.ylabel("value")
    plt.title(f"Window example â {class_name}")
    plt.legend(loc='upper right', bbox_to_anchor=(1.12,1))
    plt.grid(True)
    plt.show()

for cls, name in [(0,"normal"), (1,"pre-fall"), (2,"fall")]:
    idxs = np.where(y==cls)[0]
    if len(idxs) == 0:
        print(f"No examples for {name}")
        continue
    i = random.choice(idxs)
    print(f"Example window for {name} (meta={meta[i]})")
    plot_window_example(X[i], class_name=name)

# -----------------------
# 4) Test-set predictions, classification report, confusion matrices
# -----------------------
# ensure X_test, y_test exist
if 'X_test' not in globals() or 'y_test' not in globals():
    raise RuntimeError("X_test/y_test not found. Please re-run the train/test split cell before running this visualization cell.")

y_pred_prob = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_prob, axis=1)

print("Test accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=["normal","pre-fall","fall"], zero_division=0))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='magma', xticklabels=["normal","pre-fall","fall"], yticklabels=["normal","pre-fall","fall"])
plt.title("Confusion Matrix (counts)")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()

cm_norm = cm.astype('float') / (cm.sum(axis=1, keepdims=True) + 1e-9)
plt.figure(figsize=(6,5))
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='crest', xticklabels=["normal","pre-fall","fall"], yticklabels=["normal","pre-fall","fall"])
plt.title("Confusion Matrix (normalized)")
plt.show()

# timeline plot of predicted classes on test set (visual)
plt.figure(figsize=(12,2))
plt.imshow([y_pred], aspect='auto', cmap='viridis', extent=[0, len(y_pred), 0, 1])
plt.title("Predicted class timeline (test windows)")
plt.yticks([]); plt.xlabel("Window index (time progression)")
plt.colorbar(label="Class (0=normal,1=pre-fall,2=fall)")
plt.show()

# -----------------------
# 5) Probability trajectory for a fall trial with alerts (consecutive-window)
# -----------------------
def simulate_stream_from_trial_array(trial_array, trial_name="trial", fs=200, pref_thresh=0.45, fall_thresh=0.65, consecutive=3, visualize=True):
    wins = create_windows(trial_array, window_size=WINDOW, step_size=STEP, apply_filter=True)
    if wins.shape[0] == 0:
        print("No windows to simulate.")
        return
    times = np.arange(wins.shape[0]) * (STEP / fs)
    probs = []
    pre_count = 0; fall_count = 0
    for i, w in enumerate(wins):
        w_scaled = scaler.transform(w.reshape(1, -1)).reshape(1, WINDOW, N_CH)
        p = model.predict(w_scaled, verbose=0)[0]
        probs.append(p)
        if p[2] >= fall_thresh:
            fall_count += 1; pre_count = 0
            if fall_count >= consecutive:
                print(f"[ALERT] FALL at ~{times[i]:.2f}s  p_fall={p[2]:.3f}")
        elif p[1] >= pref_thresh:
            pre_count += 1; fall_count = 0
            if pre_count >= consecutive:
                print(f"[WARNING] PRE-FALL at ~{times[i]:.2f}s  p_prefall={p[1]:.3f}")
        else:
            pre_count = 0; fall_count = 0

    probs = np.array(probs)
    if visualize:
        plt.figure(figsize=(12,4))
        plt.plot(times, probs[:,0], label="Normal", lw=2)
        plt.plot(times, probs[:,1], label="Pre-Fall", lw=2)
        plt.plot(times, probs[:,2], label="Fall", lw=2)
        plt.xlabel("Time (s)"); plt.ylabel("Probability")
        plt.title(f"Predicted probabilities â {trial_name}")
        plt.legend(); plt.grid(True); plt.show()
    return probs, times

if len(fall_trials) > 0:
    print("Simulating and visualizing for:", fall_trials[0]["fname"])
    simulate_stream_from_trial_array(fall_trials[0]["data"], trial_name=fall_trials[0]["fname"])
else:
    print("No fall trials available to visualize.")

# -----------------------
# 6) SHAP explainability (time x channel heatmap + channel summary)
# -----------------------
# Install shap if not present (Kaggle typically has shap; this is a safe check)
try:
    import shap
except Exception:
    print("Installing shap...")
    !pip install -q shap
    import shap

# Prep flattened versions for SHAP background / wrapper
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

def predict_flat(x_flat):
    """x_flat shape (n_samples, WINDOW*nch) -> returns model probabilities"""
    x = x_flat.reshape((-1, WINDOW, N_CH))
    return model.predict(x, verbose=0)

# Choose a small background and a small number of test samples to explain
bg_size = min(30, X_train_flat.shape[0])
bg_idx = np.random.choice(X_train_flat.shape[0], bg_size, replace=False)
background = X_train_flat[bg_idx]

n_explain = min(2, X_test_flat.shape[0])
ex_idx = np.random.choice(X_test_flat.shape[0], n_explain, replace=False)
samples = X_test_flat[ex_idx]

print(f"Running SHAP: background {background.shape[0]} samples, explaining {samples.shape[0]} windows.")

shap_values = None
use_deep = False
try:
    # DeepExplainer expects model and background in original (WINDOW,nch) shape
    de = shap.DeepExplainer(model, background.reshape((-1, WINDOW, N_CH)))
    shap_vals_list = de.shap_values(samples.reshape((-1, WINDOW, N_CH)))
    shap_values = shap_vals_list  # list per class
    use_deep = True
    print("DeepExplainer succeeded.")
except Exception as e:
    print("DeepExplainer failed:", e)
    print("Falling back to KernelExplainer (slower).")
    # KernelExplainer for each class separately (use small nsamples)
    shap_values = {}
    # use kmeans summary for background to speed up
    bg_k = shap.kmeans(background, min(10, background.shape[0]))
    for class_idx in [1, 2]:  # pre-fall and fall classes
        f = lambda x: predict_flat(x)[:, class_idx]
        ke = shap.KernelExplainer(f, bg_k)
        svals = ke.shap_values(samples, nsamples=200)  # reduce nsamples if slow
        shap_values[class_idx] = svals
    print("Kernel SHAP done for classes 1 and 2.")

# plotting helpers
def plot_shap_heatmap_flat(shap_flat, sample_flat, title="SHAP heatmap", sample_i=0):
    arr = np.array(shap_flat)
    # handle DeepExplainer shapes:
    # if arr ndim==3 (n_samples, WINDOW, nch) -> pick sample_i
    if arr.ndim == 3:
        arr2 = arr[sample_i]
    elif arr.ndim == 2 and arr.shape[0] == sample_flat.shape[0] and arr.shape[1] == WINDOW*N_CH:
        arr2 = arr[sample_i].reshape(WINDOW, N_CH)
    elif arr.ndim == 1 and arr.size == WINDOW*N_CH:
        arr2 = arr.reshape(WINDOW, N_CH)
    else:
        # try reshape last dim
        try:
            arr2 = arr.reshape(sample_flat.shape[0], WINDOW, N_CH)[sample_i]
        except Exception as e:
            print("Unexpected SHAP shape:", arr.shape, "error:", e)
            return

    plt.figure(figsize=(10,4))
    sns.heatmap(arr2.T, cmap="coolwarm", center=0, cbar_kws={'label': 'SHAP value'},
                xticklabels=np.round(np.linspace(0, (WINDOW-1)/WINDOW, WINDOW),2),
                yticklabels=[f"ch{c}" for c in range(N_CH)])
    plt.xlabel("Window time (normalized)"); plt.ylabel("Channels")
    plt.title(title)
    plt.show()

def plot_channel_summary_from_shap(arr2, title="Channel importance"):
    # arr2 shape (WINDOW, N_CH)
    ch_imp = np.sum(np.abs(arr2), axis=0)
    plt.figure(figsize=(8,3))
    sns.barplot(x=[f"ch{c}" for c in range(N_CH)], y=ch_imp)
    plt.title(title); plt.ylabel("Sum |SHAP|"); plt.xticks(rotation=45)
    plt.show()

# visualize SHAP
if use_deep and isinstance(shap_values, list):
    # shap_values is list per class (len=num_classes), each element maybe (n_samples, WINDOW, N_CH)
    for class_idx, sval in enumerate(shap_values):
        print("SHAP visualization for class", class_idx)
        # plot first explained sample
        try:
            plot_shap_heatmap_flat(sval, samples, title=f"SHAP heatmap class {class_idx}", sample_i=0)
            # obtain arr2 (WINDOW,N_CH) for sample0
            arr2 = sval[0] if sval.ndim==3 else np.array(sval).reshape(samples.shape[0], WINDOW, N_CH)[0]
            plot_channel_summary_from_shap(arr2, title=f"Channel importance (class {class_idx})")
        except Exception as e:
            print("Error plotting SHAP for class", class_idx, e)
else:
    # Kernel SHAP dict keyed by class_idx
    for cls_idx, sval in shap_values.items():
        print("SHAP visualization for class", cls_idx)
        try:
            # sval often is list per sample -> pick first
            arr_flat = sval[0]
            plot_shap_heatmap_flat(np.array(arr_flat), samples, title=f"Kernel SHAP heatmap class {cls_idx}", sample_i=0)
            arr2 = np.array(arr_flat).reshape(WINDOW, N_CH)
            plot_channel_summary_from_shap(arr2, title=f"Channel importance (class {cls_idx})")
        except Exception as e:
            print("Error plotting Kernel SHAP for class", cls_idx, e)

print("All visualizations done.")
```

```python
# ====== Robust SHAP fallback: try GradientExplainer, else build surrogate + TreeExplainer ======
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import shap
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,5)

# Requirements: model, X_train, X_test, X, y, WINDOW, N_CH, scaler must exist in namespace
if 'model' not in globals():
    raise RuntimeError("model not found in environment. Load it first.")
if 'X_train' not in globals() or 'X_test' not in globals():
    raise RuntimeError("X_train/X_test not found. Run your preprocessing cell first.")

print("Input shapes:", "X_train", X_train.shape, "X_test", X_test.shape)
print("WINDOW:", WINDOW, "N_CH:", N_CH)

# Choose background and samples to explain
bg_size = min(50, X_train.shape[0])
bg_idx = np.random.choice(X_train.shape[0], bg_size, replace=False)
background_full = X_train[bg_idx]                       # shape (bg_size, WINDOW, N_CH)

n_explain = min(2, X_test.shape[0])
ex_idx = np.random.choice(X_test.shape[0], n_explain, replace=False)
samples_full = X_test[ex_idx]                           # shape (n_explain, WINDOW, N_CH)

print("Background samples:", background_full.shape, "Explaining:", samples_full.shape)

# Helper: predict probabilities for full windows
def model_predict_probs(windows_full):
    """windows_full: (n, WINDOW, N_CH) -> returns (n, n_classes)"""
    return model.predict(windows_full, verbose=0)

# First try GradientExplainer (works with TF2 models)
try:
    print("Trying shap.GradientExplainer on full model...")
    ge = shap.GradientExplainer((model), background_full)
    # explain for each class: this returns list of arrays per output (n_classes)
    shap_vals_grad = ge.shap_values(samples_full)  # list per class
    print("GradientExplainer succeeded.")
    explainer_kind = "gradient"
except Exception as e:
    print("GradientExplainer failed:", e)
    explainer_kind = "failed"

# Fallback: surrogate + TreeExplainer on downsampled windows
if explainer_kind == "failed":
    print("Building downsampled surrogate because direct gradient explainer failed.")
    # 1) Downsample factor (choose so features << bg_size)
    ds_factor = 10   # 200 -> 20 time bins; adjust if needed
    assert WINDOW % ds_factor == 0, "Choose ds_factor that divides WINDOW"
    W_ds = WINDOW // ds_factor
    print(f"Downsampling windows by factor {ds_factor} -> W_ds={W_ds}")

    def downsample_windows(windows, factor=ds_factor):
        # windows shape (n, WINDOW, N_CH)
        n = windows.shape[0]
        w = windows.reshape(n, W_ds, factor, N_CH).mean(axis=2)  # (n, W_ds, N_CH)
        return w

    # Build downsampled train background and samples
    background_down = downsample_windows(background_full)     # (bg_size, W_ds, N_CH)
    samples_down = downsample_windows(samples_full)           # (n_explain, W_ds, N_CH)

    # Flatten for surrogate
    Xbg = background_down.reshape(background_down.shape[0], -1)   # (bg_size, W_ds*N_CH)
    Xsamples = samples_down.reshape(samples_down.shape[0], -1)

    # Get target probabilities from original model for the background
    ybg_probs = model_predict_probs(background_full)             # (bg_size, n_classes)

    n_classes = ybg_probs.shape[1]
    print("Training RandomForest regressors to mimic model outputs (per-class).")
    rf_regressors = []
    for cls in range(n_classes):
        rf = RandomForestRegressor(n_estimators=100, max_depth=8, random_state=SEED, n_jobs=-1)
        rf.fit(Xbg, ybg_probs[:, cls])
        rf_regressors.append(rf)
        print(f"  Trained RF for class {cls}")

    # Now create a surrogate prediction function (on downsampled flattened inputs):
    def surrogate_predict_proba_flat(X_flat):
        # X_flat shape (n, W_ds * N_CH)
        probs = np.vstack([rf.predict(X_flat) for rf in rf_regressors]).T
        return probs

    # Use shap.TreeExplainer on each RF (TreeExplainer supports tree models directly)
    shap_values_surrogate = {}
    for cls, rf in enumerate(rf_regressors):
        print(f"Creating TreeExplainer for surrogate class {cls}")
        te = shap.TreeExplainer(rf)
        # explain the provided samples
        shap_vals = te.shap_values(Xsamples)   # shape (n_explain, W_ds*N_CH)
        shap_values_surrogate[cls] = shap_vals
    explainer_kind = "surrogate"
    print("Surrogate SHAP computed.")

# -----------------------
# Plotting SHAP heatmaps depending on path
# -----------------------
def plot_heatmap_from_shap_flat(shap_flat, sample_flat, W_time, nch, title="SHAP heatmap"):
    """
    shap_flat: either (W_time*nch,) or (n_samples, W_time*nch) or (n_samples, W_time, nch)
    sample_flat: flattened sample (for shape context) or not used
    """
    arr = np.array(shap_flat)
    # normalize shapes and pick first sample if necessary
    if arr.ndim == 3:                        # (n_samples, W_time, nch)
        arr2 = arr[0]
    elif arr.ndim == 2 and arr.shape[0] == sample_flat.shape[0] and arr.shape[1] == W_time*nch:
        arr2 = arr[0].reshape(W_time, nch)
    elif arr.ndim == 2 and arr.shape[0] == W_time and arr.shape[1] == nch:
        arr2 = arr
    elif arr.ndim == 1 and arr.size == W_time * nch:
        arr2 = arr.reshape(W_time, nch)
    else:
        try:
            arr2 = arr.reshape(-1, W_time, nch)[0]
        except Exception as e:
            print("Cannot reshape SHAP array, shape:", arr.shape, "err:", e)
            return

    plt.figure(figsize=(10,4))
    sns.heatmap(arr2.T, cmap="coolwarm", center=0,
                xticklabels=np.round(np.linspace(0, (W_time-1)/W_time, W_time),2),
                yticklabels=[f"ch{c}" for c in range(nch)])
    plt.xlabel("Time (normalized)"); plt.ylabel("Channels"); plt.title(title)
    plt.show()

def plot_channel_summary(arr2, title="Channel importance"):
    ch_imp = np.sum(np.abs(arr2), axis=0)
    plt.figure(figsize=(8,3))
    sns.barplot(x=[f"ch{c}" for c in range(arr2.shape[1])], y=ch_imp)
    plt.title(title); plt.ylabel("Sum |SHAP|"); plt.xticks(rotation=45)
    plt.show()

if explainer_kind == "gradient":
    print("Plotting SHAP from GradientExplainer (full-resolution)...")
    # shap_vals_grad is a list per class; each element often shape (n_explain, WINDOW, N_CH)
    for cls_idx, cls_shap in enumerate(shap_vals_grad):
        print("Class", cls_idx, "SHAP shape:", np.array(cls_shap).shape)
        plot_heatmap_from_shap_flat(cls_shap, samples_full.reshape(samples_full.shape[0], -1), WINDOW, N_CH, title=f"Gradient SHAP â class {cls_idx}")
        # pick arr2 for channel summary (attempt to extract)
        arr2 = None
        try:
            arr2 = cls_shap[0] if np.array(cls_shap).ndim == 3 else np.array(cls_shap).reshape(samples_full.shape[0], WINDOW, N_CH)[0]
            plot_channel_summary(arr2, title=f"Channel importance class {cls_idx}")
        except Exception:
            pass

elif explainer_kind == "surrogate":
    print("Plotting SHAP from surrogate TreeExplainer (downsampled).")
    # W_ds and factor defined earlier
    for cls_idx, sval in shap_values_surrogate.items():
        # sval shape typically (n_explain, W_ds*N_CH)
        print("Surrogate class", cls_idx, "shap shape:", np.array(sval).shape)
        # plot heatmap at downsampled resolution
        plot_heatmap_from_shap_flat(sval, Xsamples, W_ds, N_CH, title=f"Surrogate SHAP (downsampled) â class {cls_idx}")
        # channel summary
        arr2 = np.array(sval[0]).reshape(W_ds, N_CH)
        plot_channel_summary(arr2, title=f"Channel importance (surrogate) class {cls_idx}")

    # Optionally upsample heatmap to original time resolution for visual alignment:
    def upsample_heatmap(arr_ds, factor=ds_factor):
        # arr_ds shape (W_ds, N_CH) -> upsample to (WINDOW, N_CH)
        return np.repeat(arr_ds, factor, axis=0)

    # Example: show upsampled SHAP heatmap for class 1 (prefall)
    if 1 in shap_values_surrogate:
        arr_ds = np.array(shap_values_surrogate[1][0]).reshape(W_ds, N_CH)
        arr_up = upsample_heatmap(arr_ds)
        plt.figure(figsize=(10,4))
        sns.heatmap(arr_up.T, cmap="coolwarm", center=0, xticklabels=50, yticklabels=[f"ch{c}" for c in range(N_CH)])
        plt.title("Upsampled surrogate SHAP heatmap (class 1)"); plt.xlabel("Time (upsampled)"); plt.show()

print("SHAP fallback complete.")
```

```python
# ====================== FALL PREDICTION ON NEW FILE ======================
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle, re, os
from tensorflow.keras.models import load_model
from scipy.signal import butter, filtfilt
from sklearn.preprocessing import StandardScaler
import tensorflow as tf

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,5)

# -------------------------------------------------
# 1ï¸â£ Load saved model and scaler (with safe fallback)
# -------------------------------------------------
MODEL_NAME = "/kaggle/working/bilstm_attention_prefall.h5"
SCALER_NAME = "/kaggle/working/scaler_prefall.pkl"

# Define focal loss (used during training)
def focal_loss(gamma=2.0, alpha=[0.5, 1.0, 1.5]):
    alpha = tf.constant(alpha, dtype=tf.float32)
    def loss_fn(y_true, y_pred):
        y_true_flat = tf.cast(tf.reshape(y_true, (-1,)), tf.int32)
        y_true_o = tf.one_hot(y_true_flat, depth=tf.shape(y_pred)[-1])
        y_pred = tf.clip_by_value(y_pred, 1e-9, 1.0)
        p_t = tf.reduce_sum(y_true_o * y_pred, axis=-1)
        alpha_factor = tf.reduce_sum(y_true_o * alpha, axis=-1)
        modulating_factor = tf.pow(1.0 - p_t, gamma)
        loss = - alpha_factor * modulating_factor * tf.math.log(p_t)
        return tf.reduce_mean(loss)
    return loss_fn

# Safe model loading
try:
    model = load_model(MODEL_NAME, custom_objects={"focal_loss": focal_loss})
    print("â Model loaded successfully with focal_loss.")
except Exception as e:
    print("â ï¸ Direct load with focal_loss failed:", e)
    print("Attempting to load model without compilation...")
    model = load_model(MODEL_NAME, compile=False)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=3e-4),
        loss=focal_loss(gamma=2.0, alpha=[0.5, 1.0, 1.5]),
        metrics=["accuracy"]
    )
    print("â Model loaded and recompiled manually.")

# Load scaler
with open(SCALER_NAME, "rb") as f:
    scaler = pickle.load(f)
print("â Scaler loaded successfully.")

# -------------------------------------------------
# 2ï¸â£ User Input â specify SisFall text file path
# -------------------------------------------------
print("\nð Enter or paste the full path of the SisFall .txt file to test:")
file_path = input("File path: ").strip()

if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found: {file_path}")
print(f"ð Selected file: {file_path}")

# -------------------------------------------------
# 3ï¸â£ Preprocessing utilities (same as training)
# -------------------------------------------------
def parse_txt_file(path):
    rows = []
    with open(path, "r", errors="ignore") as f:
        for line in f:
            line = line.strip().rstrip(";")
            if not line:
                continue
            toks = [t.strip() for t in line.split(",") if t.strip() != ""]
            if len(toks) != 9:
                continue
            try:
                vals = [float(x) for x in toks]
            except:
                continue
            rows.append(vals)
    if len(rows) == 0:
        return np.empty((0,0), dtype=np.float32)
    return np.array(rows, dtype=np.float32)

def butter_lowpass_filter(data, fs=200, cutoff=30, order=4):
    if data.shape[0] < (order * 3):
        return data
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    try:
        return filtfilt(b, a, data, axis=0)
    except Exception:
        return data

def create_windows(trial_array, window_size=200, step_size=100, fs=200, apply_filter=True):
    if apply_filter:
        trial_array = butter_lowpass_filter(trial_array, fs=fs, cutoff=30)
    n = trial_array.shape[0]
    windows = []
    for start in range(0, n - window_size + 1, step_size):
        windows.append(trial_array[start:start + window_size])
    if len(windows) == 0:
        return np.empty((0, window_size, trial_array.shape[1]))
    return np.stack(windows, axis=0)

# -------------------------------------------------
# 4ï¸â£ Prepare the test file for prediction
# -------------------------------------------------
WINDOW, STEP = 200, 100

trial_data = parse_txt_file(file_path)
if trial_data.shape[0] == 0:
    raise ValueError("â Empty or unreadable file.")

wins = create_windows(trial_data, window_size=WINDOW, step_size=STEP, apply_filter=True)
print(f"ð Created {wins.shape[0]} windows from the file.")

# Scale with loaded scaler
nch = wins.shape[2]
wins_flat = wins.reshape(wins.shape[0], -1)
wins_scaled = scaler.transform(wins_flat).reshape(wins.shape[0], WINDOW, nch)

# -------------------------------------------------
# 5ï¸â£ Run model prediction
# -------------------------------------------------
pred_probs = model.predict(wins_scaled, verbose=0)
pred_classes = np.argmax(pred_probs, axis=1)
label_map = {0: "Normal", 1: "Pre-Fall", 2: "Fall"}

# -------------------------------------------------
# 6ï¸â£ Visualize results
# -------------------------------------------------
plt.figure(figsize=(12,4))
plt.plot(pred_probs[:,0], label="Normal", lw=2)
plt.plot(pred_probs[:,1], label="Pre-Fall", lw=2)
plt.plot(pred_probs[:,2], label="Fall", lw=2)
plt.xlabel("Window Index (time progression)")
plt.ylabel("Probability")
plt.title("Predicted Probabilities Across Windows")
plt.legend()
plt.grid(True)
plt.show()

# Predicted timeline
plt.figure(figsize=(12,2))
plt.imshow([pred_classes], aspect='auto', cmap='viridis',
            extent=[0, len(pred_classes), 0, 1])
plt.yticks([]); plt.xlabel("Window Index")
plt.title("Predicted Class Timeline (0=Normal,1=Pre-Fall,2=Fall)")
plt.show()

# -------------------------------------------------
# 7ï¸â£ Print textual summary
# -------------------------------------------------
unique, counts = np.unique(pred_classes, return_counts=True)
summary = dict(zip([label_map[k] for k in unique], counts))
print("\nð§© Prediction Summary:")
for k,v in summary.items():
    print(f"  {k:<10}: {v} windows")

if 2 in pred_classes:
    print("\nð¨ Detected potential FALL sequence in this file!")
elif 1 in pred_classes:
    print("\nâ ï¸ Pre-Fall motion detected â caution advised.")
else:
    print("\nâ No fall or pre-fall detected. Activity appears normal.")

print("\nInference complete.")
```
